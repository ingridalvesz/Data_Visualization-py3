```python
#Importando Biliotecas
import pandas as pd
import numpy as np
from IPython.display import display, HTML
```

```python
#lendo arquivo csv para analizar dados
df = pd.read_csv('relatorio_vendas.csv')
df
```

```python
#Checando informações dos dados
df.info()
```

```python
#Como a data estava como tipo errado transformei em um datetime64[ns]
df['data_pedido'] = pd.to_datetime(df['data_pedido'])
```

```python
#Checando se o tipo está correto
df.info()
```

# Demandas de Vendas

```python
#Estruturação dos dataframe 
#Criando um ranque de 10 clientes os quais mais vendemos
df_clientes_vendas = df.groupby(['nome_cliente'])['vendas'].sum().nlargest(10).copy()
df_clientes_vendas = df_clientes_vendas.reset_index()
df_clientes_vendas 
```

```python
#Aqui está uma melhor forma de veriricar o nosso ranque com os 10 clientes os quais mais vendemos
df_clientes_vendas.columns = ['Clientes', 'Vendas']
df_clientes_vendas['Ranque'] = df_clientes_vendas.index + 1
df_clientes_vendas.set_index('Ranque', inplace=True)
df_clientes_vendas
```

```python
#transformando o nosso dataframe em um objeto style
s = df_clientes_vendas.style
s
```

```python
#formatando a coluna vendas para uma melhor vizualiação
s.format({'Vendas' : 'R$ {:,.2f}'})
s
```

# Demandas Comercial

```python
#Criando um dataframe para organização	
df_vendas_lucro = df.groupby(['tipo_produto']) [['vendas', 'lucro']].sum()
df_vendas_lucro.index.name = 'Tipos de Produto'
df_vendas_lucro
```

```python
#transformando o nosso dataframe em um objeto style
estilo_produto = df_vendas_lucro.style
estilo_produto
```

```python
#Alteração do backgroud para uma melhor vizualiação
estilo_produto.format('R$ {:,.2f}')\
              .highlight_max(color='lightgreen')\
              .highlight_min(color='#F16165')
estilo_produto
```

```python
estilo_produto.to_html('dataframe_estilizado.html')
```

```python
#Adcionando gradiente
estilo_produto.format('R$ {:,.2f}').background_gradient(cmap='Greens')
estilo_produto
```

```python
estilo_produto.to_html('dataframe_estilizado2.html')
```

```python
#Alteração de cabeçalho 
cabecalho = {
    'selector':'th' ,
        'props':'font-weight: bold; font-family: Arial; text-align: center; text-transform: capitalize;'
}
estilo_produto.set_table_styles([cabecalho], overwrite=False) 
```

```python
estilo_produto.to_html('dataframe_estilizado3.html')
```

# Demanda de logística


```python
#Criando um dataframe para região
df_regiao = pd.DataFrame(df['regiao'].value_counts())
df_regiao.columns = ['N° de pedidos']
df_regiao
```

```python
#Aleterando o index para o nome correto
df_regiao.index.name = 'Região'
```

```python
#Criando uma porcentagem para pedidos por região
porcentagem = df_regiao['N° de pedidos'].to_numpy()
porcentagem = 100 * porcentagem/porcentagem.sum()
porcentagem
```

```python
#Condicionando o dataframe com porcentagem de pedidos feitos por cada região
df_regiao['Porcentagem Região'] = porcentagem
df_regiao
```

```python
#Criando o objeto style
estilo_regiao = df_regiao.style
estilo_regiao
```

```python
#Fazendo a padronização	do cabeçalho com CSS
cabecalho = {
    'selector': 'th',
    'props': 'font-weight: bold; font-family: Arial; text-align: right; background-color: white'
}
celulas = {
    'selector': 'td',
    'props': 'background-color: white;'
}

estilo_regiao.set_table_styles([cabecalho,celulas])
```

```python
estilo_regiao.format({'Porcentagem Região':'{:.2f} %'})\
             .bar(subset='Porcentagem Região', vmin=0, vmax=100.0, color='#9CD33B')
```
